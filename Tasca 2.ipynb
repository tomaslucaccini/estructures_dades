# Ejercicio 3:

```python
lista_desordenada = [3, 4, 5, 1, 3, 3, 9, 2, 10, 83, 2, 4, 6]
lista_desordenada
```

```python
# cantidad de números dentro de la lista

len(lista_desordenada)
```

```python
# Cuántas veces aparece el número 3?

lista_desordenada.count(3)

```

```python
# Contar "3" y "4"

contar1 = 0
contar2 = 0

for x in lista_desordenada:
    if x == 3:
        contar1 += 1
    if x == 4:
        contar2 += 1

cuenta_total = contar1 + contar2
cuenta_total
```

```python
# Número más grande

lista_desordenada.sort()
lista_desordenada[-1]
```

```python
# Tres números más pequeños --> ordenar ascendente y seleccionar primeros tres:

lista_desordenada.sort()
lista_desordenada[0:3]
```

```python
# Mínimo & máximo para conocer el Rango. 

min_val = min(lista_desordenada)
max_val = max(lista_desordenada)

print("El valor máximo es ", max_val, "y el valor mínimo es ", min_val)

rango = max_val - min_val
print("El rango de la lista es ", rango)
```
